```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20250526.csv')
```

```python
sns.set(style="whitegrid")
```

```python
print("\nColumns:", df.columns)
print("\nDataset Info:")
print(df.info())
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

df = df.sort_values('Date')
```

```python
df_melted = df.melt(id_vars='Date',
                    value_vars=['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School', 'Other'],
                    var_name='Service',
                    value_name='Journeys')
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_melted, x='Date', y='Journeys', hue='Service')
plt.title("Daily Passenger Journeys by Service Type")
plt.xlabel("Date")
plt.ylabel("Number of Journeys")
plt.legend(title='Service Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_journeys = df_melted.groupby("Service")["Journeys"].mean().sort_values(ascending=False)
print("\nAverage Daily Journeys by Service Type:\n", avg_journeys)
```

```python
total_daily = df_melted.groupby("Date")["Journeys"].sum().reset_index()
top_days = total_daily.sort_values("Journeys", ascending=False).head(5)
print("\nTop 5 Busiest Overall Days:\n", top_days)
```

```python
plt.figure(figsize=(14, 4))
plt.plot(total_daily['Date'], total_daily['Journeys'], label="Total Journeys", color="purple")
plt.title("Overall Daily Public Transport Journeys")
plt.xlabel("Date")
plt.ylabel("Journeys")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
url = "https://www.data.act.gov.au/api/views/nkxy-abdj/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df = df.sort_values('Date')
```

```python
def forecast_service(service_name):
    print(f"\n--- Forecasting: {service_name} ---")


    data = df[['Date', service_name]].rename(columns={"Date": "ds", service_name: "y"})


    model = Prophet()
    model.fit(data)

    # Create future 7 days
    future = model.make_future_dataframe(periods=7)
    forecast = model.predict(future)

    # Plot forecast
    fig = model.plot(forecast)
    plt.title(f"{service_name} - Forecast for Next 7 Days")
    plt.xlabel("Date")
    plt.ylabel("Passenger Journeys")
    plt.tight_layout()
    plt.show()

    # Return the 7-day forecast
    return forecast[['ds', 'yhat']].tail(7)
```

```python
services = ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School']
forecasts = {}

for service in services:
    forecasts[service] = forecast_service(service)
```
